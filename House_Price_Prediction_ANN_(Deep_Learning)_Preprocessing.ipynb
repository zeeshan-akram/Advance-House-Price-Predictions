```
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
import keras
from keras.wrappers.scikit_learn import KerasRegressor
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import LeakyReLU, PReLU, ELU
from keras.layers import Dropout
from keras.callbacks import ModelCheckpoint
```

```
test_data = pd.read_csv('/content/drive/My Drive/Deep learning/Artificial Neural Network/House Price/final_test_data_model_deep_learning.csv')
train_data = pd.read_csv('/content/drive/My Drive/Deep learning/Artificial Neural Network/House Price/final_train_data_model_deep_learning.csv')
```

```
inputs = train_data.drop('PriceSales', axis=1)
outputs = train_data['PriceSales']
```

```
x_train, x_test, y_train, y_test = train_test_split(inputs, outputs, test_size=0.2, random_state=42)
```

```
scale = MinMaxScaler()
```

```
x_train_s = scale.fit_transform(x_train)
y_train_s = scale.fit_transform(y_train.values.reshape(-1,1))
x_test_s = scale.fit_transform(x_test)
y_test_s = scale.fit_transform(y_test.values.reshape(-1, 1))
```

```
regressor = Sequential()
#hidden and input layer
regressor.add(Dense(units=20, kernel_initializer='he_normal', activation='relu', input_dim=102))

#hidden layers
regressor.add(Dense(units=15, kernel_initializer='he_normal', activation='relu'))
regressor.add(Dropout(0.2))
regressor.add(Dense(units=10, kernel_initializer='he_normal', activation='relu'))
#regressor.add(Dropout(0.2))

#output layer
regressor.add(Dense(units = 1, kernel_initializer='he_normal', activation='relu'))

#compile
regressor.compile(loss='mean_squared_error', optimizer='adam', metrics=[keras.metrics.RootMeanSquaredError()])
```

```
regressor.summary()
```

```
regressor_history = regressor.fit(x_train_s, y_train_s, validation_split=0.2, batch_size = 32, epochs=200, verbose=2)
```

```
y_pred = regressor.predict(x_test_s)
```

```
predictions = scale.inverse_transform(y_pred)
```
