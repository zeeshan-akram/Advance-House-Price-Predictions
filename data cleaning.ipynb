Now we can merge both datasets to perform some feature engineering.<br>

## Missing Values
I'll fill missing values with the help of info provided the data provider

```python
null_values = all_data.isnull().sum()
null_values = null_values[null_values != 0]
null_values = pd.DataFrame(null_values.sort_values(ascending=False), columns=['Null Values'])
null_values['Percentage'] = round((null_values['Null Values']/all_data.shape[0])*100,2)
null_values['dtypes'] = all_data[null_values.index].dtypes
null_values
```

## Handle Categorical Missing values

#### Let's fill these missing values

PoolQC: No Pool<br>
MiscFeature: No elvatore, tennis court etc features<br>
Alley: No access to alley <br>
Fence: No fence available <br>
FireplaceQu: no FireplaceQu <br>
GarageQual: No Garage <br>
GarageFinish: No Garage <br>
GarageCond: No Garage <br>
GarageType: No Garage <br>
BsmtExposure: No Basement <br>
BsmtCond: No Basement <br>
BsmtQual: No Basement <br>
BsmtFinType2: No Basement <br>
BsmtFinType1: No Basement <br>
MasVnrType: No information available <br>
MSZoning: No information available <br>
Functional: No information available <br>
Utilities: No information available <br>
Exterior2nd: No information available <br>
Exterior1st: No information available <br>
SaleType: No information available <br>
Electrical: No information available <br>
KitchenQual: No information available 

separate features by category and numerical

```python
category_features = all_data.dtypes[all_data.dtypes =='O'].index
len(category_features)
```

```python
numerical_features = all_data.dtypes[all_data.dtypes !='O'].index
len(numerical_features)
```

```python
# filling none values which don't have these features
cat_with_none = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageQual', 'GarageFinish', 'GarageCond'
                 , 'GarageType', 'BsmtExposure', 'BsmtCond', 'BsmtQual', 'BsmtFinType2', 'BsmtFinType1']
for col in cat_with_none:
    all_data[col].fillna('None', inplace=True)
```

```python
# fill MasVnrType by grouping categories
all_data['MasVnrType'] = all_data.groupby(['MSSubClass', 'LotShape','Neighborhood', 'Exterior1st', 'Exterior2nd',
                        'ExterQual', 'Foundation'])['MasVnrType'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill remaining MasVnrType by grouping categories
all_data['MasVnrType'] = all_data.groupby(['MSSubClass', 'LotShape',
                                           'Foundation'])['MasVnrType'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill MSZoning by grouping categories
all_data['MSZoning'] = all_data.groupby(['MSSubClass', 'HouseStyle', 
                        'OverallQual', 'OverallCond'])['MSZoning'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill remaining
all_data['MSZoning'] = all_data.groupby(['MSSubClass',
                        'OverallQual'])['MSZoning'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill Utilities by grouping categories
all_data['Utilities'] = all_data.groupby(['MSSubClass',
                        'OverallQual'])['Utilities'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill Utilities by grouping categories
all_data['Functional'] = all_data.groupby(['MSSubClass', 'OverallCond'
                                          ])['Functional'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
# fill remaining
all_data['Functional'] = all_data.groupby(['OverallCond'
                                          ])['Functional'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill SaleType by grouping categories
all_data['SaleType'] = all_data.groupby(['MSSubClass', 'OverallCond',
                        'OverallQual'])['SaleType'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill KitchenQual by grouping categories
all_data['KitchenQual'] = all_data.groupby(['MSSubClass', 'OverallCond', 'KitchenAbvGr',
                        'OverallQual'])['KitchenQual'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill KitchenQual by grouping categories
all_data['Electrical'] = all_data.groupby(['MSSubClass', 'OverallCond', 'MSZoning', 'Neighborhood',
                        'OverallQual'])['Electrical'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill Exterior2nd by grouping categories
all_data['Exterior2nd'] = all_data.groupby(['MSSubClass', 'OverallCond', 'MSZoning', 'Neighborhood',
                        'OverallQual', 'HouseStyle', 'MasVnrType'])['Exterior2nd'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

```python
# fill Exterior1st by grouping categories
all_data['Exterior1st'] = all_data.groupby(['MSSubClass', 'OverallCond', 'MSZoning', 'Neighborhood',
                        'OverallQual', 'HouseStyle', 'MasVnrType'])['Exterior1st'].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))
```

## Handle Numerical Missing values

```python
# fill LotFrontage by grouping categories
all_data['LotFrontage'] = all_data.groupby(['MSSubClass', 'Neighborhood', 'Street', 'HouseStyle'])['LotFrontage'].transform(lambda x: x.fillna(x.mean()))
```

```python
# fill remaining LotFrontage by grouping categories
all_data['LotFrontage'] = all_data.groupby(['Neighborhood', 'Street'])['LotFrontage'].transform(lambda x: x.fillna(x.mean()))
```

```python
# fill remaining GarageYrBlt with zero because there is no Garage
all_data['GarageYrBlt'].fillna(0, inplace=True)
```

```python
# fill MasVnrArea by grouping categories
all_data['MasVnrArea'] = all_data.groupby(['MSSubClass', 'LotShape','Neighborhood', 'Exterior1st', 'Exterior2nd',
                        'ExterQual', 'Foundation'])['MasVnrArea'].transform(lambda x: x.fillna(x.mean()))
```

```python
# fill remaining MasVnrArea by grouping categories
all_data['MasVnrArea'] = all_data.groupby(['MSSubClass', 'LotShape',
                                           'Foundation'])['MasVnrArea'].transform(lambda x: x.fillna(x.mean()))
```

```python
# fill BsmtHalfBath with zero because there is no bsmt
all_data['BsmtHalfBath'].fillna(0, inplace=True)
```

```python
# fill BsmtFullBath with zero because there is no bsmt
all_data['BsmtFullBath'].fillna(0, inplace=True)
```

```python
# fill GarageArea with zero because there is no Garage
all_data['GarageArea'].fillna(0, inplace=True)
```

```python
# fill GarageCars with zero because there is no Garage
all_data['GarageCars'].fillna(0, inplace=True)
```

```python
# fill TotalBsmtSF with zero because there is no bsmt
all_data['TotalBsmtSF'].fillna(0, inplace=True)

# fill BsmtUnfSF with zero because there is no bsmt
all_data['BsmtUnfSF'].fillna(0, inplace=True)

# fill BsmtFinSF2 with zero because there is no bsmt
all_data['BsmtFinSF2'].fillna(0, inplace=True)

# fill BsmtFinSF1 with zero because there is no bsmt
all_data['BsmtFinSF1'].fillna(0, inplace=True)
```
